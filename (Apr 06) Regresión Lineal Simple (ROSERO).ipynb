## Explicaciones Clave
**Importación de datos:** Se asume que el archivo co2.csv contiene varias columnas, de las que se usarán dos: Engine Size(L) y CO2 Emissions(g/km).<br>
**División del dataset:** Usamos train_test_split para dividir los datos en un conjunto de entrenamiento (2/3) y prueba (1/3).<br>
**Entrenamiento del modelo:** Utilizamos LinearRegression de sklearn para ajustar un modelo lineal a los datos de entrenamiento.<br>

## Visualización de Resultados
- Gráfico para el conjunto de entrenamiento: muestra los puntos reales y la línea de regresión ajustada.
- Gráfico para el conjunto de prueba: muestra los puntos reales del conjunto de prueba y la misma línea ajustada con el conjunto de entrenamiento.
- Impresión de resultados de Multiple R para conocer el grado de correlación entre las variables.
- Impresión de resultados de R Cuadrado para saber qué tan bien predice los datos el modelo, es decir, su nivel de confiabilidad.

### Importar las bibliotecas necesarias

```python
import numpy as np  # Para operaciones matemáticas y manejo de matrices
import matplotlib.pyplot as plt  # Para visualización de datos
import pandas as pd  # Para manipulación y análisis de datos
```

### Importar el dataset

```python
dataset = pd.read_csv('co2.csv')

X = dataset[["Engine Size(L)"]].values  # Variable independiente (tamaño del motor en litros)
y = dataset["CO2 Emissions(g/km)"].values  # Variable dependiente (emisión de gramos de CO2 por kilómetro)
```

Para la Variable Independiente se utiliza doble corchete [[]] porque devuelve un arreglo de 2 dimensiones, dado que la librería scikit-learn espera una matriz de entrada para esta variable, aunque solo tenga una columna 

### Dividir el dataset en conjunto de entrenamiento y conjunto de prueba

```python
from sklearn.model_selection import train_test_split
# Usamos 2/3 para entrenamiento y 1/3 para prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=0)
```

### Entrenar el modelo de Regresión Lineal Simple con el conjunto de entrenamiento

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()  # Crear el modelo de regresión lineal
regressor.fit(X_train, y_train)  # Ajustar el modelo a los datos de entrenamiento
```

### Predecir los resultados del conjunto de prueba

```python
y_pred = regressor.predict(X_test)  # Generar predicciones para los datos de prueba
```

### Visualizar los resultados del conjunto de entrenamiento

```python
plt.scatter(X_train, y_train, color='green')  # Puntos reales en el conjunto de entrenamiento
plt.plot(X_train, regressor.predict(X_train), color='red')  # Línea de regresión ajustada
plt.title('Emisiones CO2 vs Tamaño del Motor (Conjunto de Entrenamiento)')  # Título del gráfico
plt.xlabel('Engine Size (L)')  # Etiqueta del eje X
plt.ylabel('CO2 Emissions(g/km)')  # Etiqueta del eje Y
plt.show()  # Mostrar el gráfico
```

### Visualizar los resultados del conjunto de prueba

```python
plt.scatter(X_test, y_test, color='salmon')  # Puntos reales en el conjunto de prueba
plt.plot(X_train, regressor.predict(X_train), color='blue')  # Línea de regresión ajustada (misma que en entrenamiento)
plt.title('Emisiones CO2 vs Tamaño del Motor (Conjunto de Prueba)')  # Título del gráfico
plt.xlabel('Engine Size (L)')  # Etiqueta del eje X
plt.ylabel('CO2 Emissions(g/km)')  # Etiqueta del eje Y
plt.show()  # Mostrar el gráfico
```
